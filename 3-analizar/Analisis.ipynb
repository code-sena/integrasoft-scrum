```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo de Excel
df = pd.read_excel('cuestionario_respuestas.xlsx')

# Graficar la distribucin de respuestas para "Le resulta fcil navegar por las categoras de productos?"
plt.figure(figsize=(8,6))
df['NavegacionFacil'].value_counts().plot(kind='bar', color=['#86bf8a', '#6a4c93', '#ffd166'])
plt.title('Facilidad de Navegacin en la Tienda en Lnea')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Qu tan importante es para usted que el sistema de inventario est siempre actualizado?"
plt.figure(figsize=(8,6))
df['ImportanciaInventario'].value_counts().plot(kind='bar', color=['#ffb4a2', '#ff677d', '#d4a5a5'])
plt.title('Importancia del Control de Inventario')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Le gustara que el sistema le informara sobre la disponibilidad de productos en tiempo real?"
plt.figure(figsize=(8,6))
df['NotificacionDisponibilidad'].value_counts().plot(kind='bar', color=['#5e4fa2', '#fbc2eb', '#6a4c93'])
plt.title('Preferencia por Notificacin de Disponibilidad')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Qu tan til le resulta recibir notificaciones sobre el estado de su pedido?"
plt.figure(figsize=(8,6))
df['UtilidadNotificacionesEstado'].value_counts().plot(kind='bar', color=['#98c9e2', '#ff9b84', '#f6d6ad'])
plt.title('Utilidad de Notificaciones de Estado del Pedido')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Mtodo de pago preferido"
plt.figure(figsize=(8,6))
df['MetodoPagoPreferido'].value_counts().plot(kind='bar', color=['#ff9a8b', '#ffb6b9', '#86bf8a', '#ffddc1'])
plt.title('Mtodos de Pago Preferidos')
plt.xlabel('Mtodo de Pago')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=45)
plt.show()

# Graficar la distribucin de respuestas para "Le gustara que la tienda ofrezca opciones para guardar su informacin de pago?"
plt.figure(figsize=(8,6))
df['GuardarInfoPago'].value_counts().plot(kind='bar', color=['#ff77a9', '#bbd8e1', '#a6a6c8'])
plt.title('Preferencia por Guardar Informacin de Pago')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Le parece til tener acceso a reportes detallados sobre el rendimiento de sus compras?"
plt.figure(figsize=(8,6))
df['UtilidadReportes'].value_counts().plot(kind='bar', color=['#ffb3c1', '#ff9a8b', '#d8bfd8'])
plt.title('Utilidad de Reportes Detallados sobre Compras')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Qu tan importante es para usted que el sitio web sea fcil de usar y accesible desde cualquier dispositivo?"
plt.figure(figsize=(8,6))
df['ImportanciaAccesibilidad'].value_counts().plot(kind='bar', color=['#f6a4b7', '#fbc2eb', '#6a4c93'])
plt.title('Importancia de la Accesibilidad del Sitio Web')
plt.xlabel('Respuesta')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=0)
plt.show()

# Graficar la distribucin de respuestas para "Qu caractersticas adicionales le gustara ver en una tienda en lnea?"
plt.figure(figsize=(8,6))
df['CaracteristicasAdicionales'].value_counts().plot(kind='bar', color=['#ffb4a2', '#fcd5ce', '#6a4c93', '#ff9b84', '#d8bfd8'])
plt.title('Caractersticas Adicionales Deseadas en la Tienda en Lnea')
plt.xlabel('Caracterstica')
plt.ylabel('Nmero de Usuarios')
plt.xticks(rotation=45)
plt.show()
```
